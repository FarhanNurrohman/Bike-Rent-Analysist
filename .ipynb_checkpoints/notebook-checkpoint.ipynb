## Exploratory Data Analysis (EDA)

Selanjutnya kita melakukan exploaratory data analysis, pertama - tama saya ingin mencari pada musim apa orang paling banyak melakukan rental sepeda dengan kode seperti dibawah:

```python
bike_df['season'] = bike_df.season.apply(lambda x : 'spring' if x == 1 else('summer' if x == 2 else('fall' if x == 3 else 'winter')))
```

```python
bike_df.drop(columns=['temp','atemp','hum','windspeed','mnth'], inplace=True)
```

```python
bike_df.groupby(by='season').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
}).sort_values(by="cnt", ascending=False)
```

Terlihat dari data di atas bahwa pada musim fall orang - orang banyak melakukan rental sepeda. Setelah melihat musim apa yang paling banyak orang melakukan rental sepeda saya ingin melihat pada waktu apa orang - orang melakukan rental sepeda dengan kode seperti dibawah:

```python
bike_df['time'] = bike_df.hr.apply(lambda x : 'Morning' if 5<=x<12 else ( 'Afternoon' if  12<=x<17 else('Evening' if 17<=x<21 else 'night')))
bike_df.sample()
bike_df.drop(columns=['hr'], inplace=True)
```

```python
bike_df.groupby(by='time').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
}).sort_values(by="cnt", ascending=False)
```

Terlihat bahwa orang - orang banyak melakukan rental sepeda pada waktu sore hari. setalah melihat waktu apa orang banyak melakukan rental sepeda saya ingin melihat pada tahun apa orang - orang banyak melakukan rental sepeda dengan kode seperti dibawah:

```python
bike_df['yr'] = bike_df.yr.apply(lambda x: 2011 if x == 0 else 2012)
```

```python
bike_df.groupby(by='yr').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
}).sort_values(by="cnt", ascending=False)
```

Terlihat bahwa oarng - orang banyak melakaukan rental pada tahun 2012, sehingga dapat diindikasikan bahwa ada kenaiakan perentah dari tahun 2011 ke tahun 2012. setelah itu saya ingin melihat apakah pada musim libur banyak melakuakan renal sepeda dengan kode seperti dibawah:

```python
bike_df['holiday'] = bike_df.holiday.apply(lambda x: 'holiday' if x==1 else 'not holiday')
```

```python
bike_df.groupby(by='holiday').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
}).sort_values(by="cnt", ascending=False)
```

Terlihat bahwa malah orang - orang melakukan renal sepeda pada musim biasa buakn pada musim libur. saya berpikir apakah orang - orang sering merental sepeda pada weekend atau hari biasa, untuk itu saya menggunakan kode seperti dibawah:

```python
bike_df['workingday'] = bike_df.workingday.apply(lambda x: 'weekend' if x==1 else 'weekday')
```

```python
bike_df.groupby(by='workingday').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
}).sort_values(by="cnt", ascending=False)
```

Terlihat bahwa pada hari - hari weekday orang - orang sering melakukan rental sepeda ketimbang hari - hari weekend sehingga dapat disipulkan bahwa bersepeda adalah sudah menjadi bagina dari gawa hidup masyarakat. Selanjutnya saya penasaran pada kondisi cuaca seperti apa orang - orang banyak merental sepeda, untuk melihat itu saya menggunakan kode seperti dibawah:

```python
bike_df['weathersit']=bike_df.weathersit.apply(lambda x: "good" if x==1 else("moderate" if x==2 else("bad" if x ==3 else "disaster")))
```

## Visualization & Explanatory Analysis

Pertama - taman untuk visualisasi data saya ingin melihat tren data berdasarkan bulan dengan kode seperti dibawah:

```python
mounthly_bike_df = bike_df.resample(rule='M', on='dteday').agg({
    'casual':'sum',
    'registered':'sum',
    "cnt": "sum"
})
mounthly_bike_df.index = mounthly_bike_df.index.strftime('%y-%b')
mounthly_bike_df = mounthly_bike_df.reset_index()
mounthly_bike_df.rename(columns={
    'casual':"total_casual_rent",
    'registered':'total_registered_rent',
    "cnt": "total_all_rent"
}, inplace=True)
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(mounthly_bike_df["dteday"], mounthly_bike_df["total_all_rent"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of Total Bike Renter (2011 - 2012)", loc="center", fontsize=20) 
plt.xticks(fontsize=10, rotation=45) 
plt.yticks(fontsize=10) 
plt.show()
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(mounthly_bike_df["dteday"], mounthly_bike_df["total_registered_rent"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of Total Registered Bike Benter (2011 - 2012)", loc="center", fontsize=20) 
plt.xticks(fontsize=10, rotation=45) 
plt.yticks(fontsize=10) 
plt.show()
```

```python
plt.figure(figsize=(10, 5)) 
plt.plot(mounthly_bike_df["dteday"], mounthly_bike_df["total_casual_rent"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of Total Casual Bike Renter (2011 - 2012)", loc="center", fontsize=20) 
plt.xticks(fontsize=10, rotation=45) 
plt.yticks(fontsize=10) 
plt.show()
```

terllihat bahwa ketida data tersebut memiliki pola yang sama yaitu akan naik hingga Mei sapai Juni kemudian turun di bulan November samapai Janurai kemudain naik lagi hingga puncaknya pada Juni sampai September kemudian turun hingga pada desember. ini wajar karena musim Panas mulai pada  bulan Juni hingga Agustus dan Gugur mulai pada Sepetember hingga November dan kedua musim itu musim yang cocok untuk bersepeda. Kemudain saya ingin memvisualisasian pada musim apa orang - orang banyak melakukan peminjaman sepeda dengan kode seperti dibawah:

```python
bike_sesional_df = bike_df.groupby(by='season').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
}).sort_values(by="cnt", ascending=False)
bike_sesional_df = bike_sesional_df.reset_index()

bike_sesional_df.rename(columns={
    'casual':"total_casual_rent",
    'registered':'total_registered_rent',
    "cnt": "total_all_rent"
}, inplace=True)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(y="total_registered_rent", x="season", data=bike_sesional_df, palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Total Registered Renter ", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
 
sns.barplot(y="total_casual_rent", x="season", data=bike_sesional_df, palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Total Casual Renter", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
 
sns.barplot(y="total_all_rent", x="season", data=bike_sesional_df, palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Total All Renter", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
 
plt.suptitle("The Season When Most People Rent Bike", fontsize=20)
plt.show()
```

Dapat dilihat bahawa orang - oarng bayak melakukan peminjama yaitu pada musim Gugur dan musim Panas. ini sama seperti diagram garis sebelumnya. kemudain saya ingin memvisualisasikan pada waktu apa orang - orang bayak melakuakan peminjaman sepeda dengan kode seperti dibawah:

```python
bike_time_df = bike_df.groupby(by='time').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
})
bike_time_df = bike_time_df.reset_index()
bike_time_df.rename(columns={
    'casual':"total_casual_rent",
    'registered':'total_registered_rent',
    "cnt": "total_all_rent"
}, inplace=True)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(y="total_registered_rent", x="time", data=bike_time_df.sort_values(by="total_registered_rent", ascending=False), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Total Registered Rent", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
 
sns.barplot(y="total_casual_rent", x="time", data=bike_time_df.sort_values(by="total_casual_rent", ascending=False), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Total Casual Rent", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
sns.barplot(y="total_all_rent", x="time", data=bike_time_df.sort_values(by="total_all_rent", ascending=False), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Total All Renter", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
 
plt.suptitle("The Time When Most People Rent Bike", fontsize=20)
plt.show()
```

Pada visualisasi daiatas dilihat bahwa orang - oarang banyak melakukan peminjaman pada waktu siang dan sore hari. kemudain saya ingin mengvisualisasikan pada bari seperti apa orang- orang mekakukan peminjaman sepeda dengan kode seperti dibawah:

```python
bike_holiday_df = bike_df.groupby(by='holiday').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
}).sort_values(by="cnt", ascending=False)

bike_holiday_df = bike_holiday_df.reset_index()
bike_holiday_df.rename(columns={
    'holiday':'workingday',
    'casual':"total_casual_rent",
    'registered':'total_registered_rent',
    "cnt": "total_all_rent"
}, inplace=True)
```

```python
bike_holiday_df = bike_holiday_df.drop(0)
```

```python
bike_weeked_df = bike_df.groupby(by='workingday').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
})
bike_weeked_df = bike_weeked_df.reset_index()
bike_weeked_df.rename(columns={
    'casual':"total_casual_rent",
    'registered':'total_registered_rent',
    "cnt": "total_all_rent"
}, inplace=True)
```

```python
new_row = {'workingday':bike_holiday_df.values[0][0], 'total_casual_rent':bike_holiday_df.values[0][1], 'total_registered_rent':bike_holiday_df.values[0][2], 'total_all_rent':bike_holiday_df.values[0][3]}
bike_weeked_df.loc[2] = new_row
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(y="total_all_rent", x="workingday", data=bike_weeked_df.sort_values(by="total_all_rent", ascending=False), palette=colors, ax=ax)
ax.set_ylabel(None)
ax.set_xlabel(None)
ax.tick_params(axis='x', labelsize=15)
 
plt.suptitle("The Day When Most People Rent Bike", fontsize=20)
plt.show()
```

Dapat terlihata bahwa orang - orang bayak menyewa sepeda pada hari kerja, ini dapat mengindikasikan bahwa penyewa kemungkianan adalah kalangan pekerja, siswa, atau mahasiswa dan jika itu benar maka bersepeda sudah menjadi bagian dari gaya hidup masyarakat. kemudain saya ingin melihat pada kondisi cuaca seperti apa orang - orang bayak melakukan penyewaan sepeda dengan kode sepeti dibawah:

```python
bike_weather_df = bike_df.groupby(by='weathersit').agg({
    'registered':'sum',
    'casual':'sum',
    'cnt':'sum'
})
bike_weather_df = bike_weather_df.reset_index()
bike_weather_df.rename(columns={
    'weathersit':'weather',
    'casual':"total_casual_rent",
    'registered':'total_registered_rent',
    "cnt": "total_all_rent"
}, inplace=True)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(y="total_all_rent", x="weather", data=bike_weather_df.sort_values(by="total_all_rent", ascending=False), palette=colors, ax=ax)
ax.set_ylabel(None)
ax.set_xlabel(None)
ax.tick_params(axis='x', labelsize=15)
 
plt.suptitle("The Weather When Most People Rent Bike", fontsize=20)
plt.show()
```

terlihat bahwa pada kondisi cuaca baguslah orang - orang banyak melakukan penyewaan sepeda. ini wajar karena pada kondisi cuaca seperti ini lah paling aman untuk bersepeda.
